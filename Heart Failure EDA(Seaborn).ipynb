# HEART FAILURE EXPLORATORY DATA ANALAYSIS

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

# DATA READING:

```python
data=pd.read_csv('heart_failure.csv')
data.head(10)
```

# Correlation Matrix

```python
corr=data.corr()
corr
```

# HEATMAP:

```python
sns.heatmap(corr,annot=False,linewidth=0.00001,cmap="YlGnBu")
```

# DATA ANALYSIS:

```python
sns.histplot(data['age'],color='skyblue',kde=True)
plt.xlabel('Number of patients')
plt.ylabel('Age')
plt.title('Distribution of Age of the number of Patients')
plt.ylim(0,100)
plt.show()
```

```python
graph=sns.histplot(data['ejection_fraction'],color='darkblue',kde=True)
plt.ylabel('Number of patients')
plt.title('Distribution of Ejection Fraction among the Patients')
plt.xlabel('Ejection Fraction(%)')
graph.axvline(55,color='darkblue',label='Normal Range')
graph.axvline(75,color='darkblue')
plt.ylim(0,100)
plt.legend()
plt.show()
```

```python
graph=sns.histplot(data['serum_creatinine'],color='magenta',kde=True)
plt.ylabel('Number of patients')
plt.xlabel('Serum Creatinine')
plt.title('Distribution of Creatinine among the Patients')
plt.ylim(0,100)
graph.axvline(0.84,color='darkmagenta',label='Normal Range')
graph.axvline(1.21,color='darkmagenta')
#plt.xlim(0,5)
plt.legend()
plt.show()
```

```python
graph=sns.histplot(data['serum_sodium'],color='maroon',kde=True)
plt.ylabel('Number of patients')
plt.xlabel('Serum Sodium')
plt.title('Distribution of Sodium among the Patients')
graph.axvline(135,color='maroon',label='Normal Range')
graph.axvline(145,color='maroon')
plt.ylim(0,100)
plt.legend()
plt.show()
```

```python
graph=sns.histplot(data['platelets'],color='green',kde=True)
plt.xlabel('Platelets')
plt.ylabel('Number of Patients')
graph.axvline(215000,color='brown',label='male range')
graph.axvline(245000,color='brown')
graph.axvline(240000,color='yellow',label='female range')
graph.axvline(265000,color='yellow')
plt.title('Distribution of Platelets among the Patients')
plt.ylim(0,60)
plt.legend()
plt.show()
```

```python
graph=sns.histplot(data['creatinine_phosphokinase'],color='black',kde=True)
plt.xlabel('Creatinine Phosphokinase')
plt.ylabel('Number of Patients')
plt.title('Distribution of Creatinine Phosphokinase among the Patients')
graph.axvline(10,color='black',label='Normal Range')
graph.axvline(120,color='black')
#plt.xlim(0,3500)
plt.ylim(0,115)
plt.legend()
plt.show()
```

```python
sns.set_style("darkgrid")
sns.jointplot(x='ejection_fraction',y='serum_sodium',data=data,hue='high_blood_pressure')
plt.xlabel('Ejection fraction')
plt.ylabel('Serum sodium')
plt.show()
```

```python
sns.set_style("darkgrid")
sns.jointplot(x='ejection_fraction',y='platelets',data=data,hue='high_blood_pressure')
plt.show()
```

```python
graph=sns.boxplot(x="high_blood_pressure", y="ejection_fraction", data=data)
plt.xlabel('High BP')
plt.ylabel('Ejection Fraction')
graph.axhline(55,color='red',label='Normal Range')
graph.axhline(75,color='red')
plt.title('Boxplot of Ejection Fraction between High and Normal Bp')
plt.legend()
plt.show()
```

```python
graph=sns.boxplot(x="anaemia", y="platelets", data=data,hue='sex')
plt.xlabel('Anaemia')
plt.ylabel('Platelets')
graph.axhline(215000,color='orange',label='male range')
graph.axhline(245000,color='orange')
graph.axhline(240000,color='blue',label='female range')
graph.axhline(265000,color='blue')
plt.title('Boxplot of Platelets distribution between Anaemic and Non-Anaemic Patients among different sex')
plt.legend()
plt.show()
```

```python
graph=sns.boxplot(x="smoking", y="ejection_fraction", data=data)
plt.ylabel('Ejection Fraction')
plt.xlabel('Smoking')
plt.title('Boxplot of Ejection Fraction between Non-Smoker and Smoker')
graph.axhline(55,color='red',label='Normal range')
graph.axhline(75,color='red')
plt.legend()
plt.show()
```

```python
graph=sns.violinplot(x="diabetes", y="serum_creatinine", data=data)
plt.ylabel('Creatinine')
plt.xlabel('Diabetes')
plt.title('Violinplot distribution of Creatinine among Non-Diabetic and Diabetic Patients ')
graph.axhline(0.84,color='red',label='Normal Range')
graph.axhline(1.21,color='red')
plt.legend()
plt.show()
```

```python
graph=sns.violinplot(x="sex", y="serum_sodium", data=data,hue='smoking',split=True)
plt.ylabel('Serum Sodium')
plt.xlabel('Sex')
plt.title('Violinplot of Serum Sodium among different sex and smoking habitats')
graph.axhline(135,color='red',label='Normal Range')
graph.axhline(145,color='red')
plt.legend()
plt.show()
```

```python
sns.jointplot(x='creatinine_phosphokinase',y='serum_creatinine',data=data,hue='diabetes')
plt.show()
```

```python
graph=sns.violinplot(x="DEATH_EVENT", y="creatinine_phosphokinase", data=data,hue='sex')
graph.axhline(120,color='black',label='Normal Range')
plt.ylabel('Creatinine Phosphokinase')
plt.xlabel('DEATH EVENT')
plt.title('Violinplot of Creatinine Phosphokinase among Death Events')
plt.legend()
plt.show()
```
